# **Question 2: Binomial Tree approximation**

We want to create a tree model that we can use to approximate the black scholes price of a given stock!

Our function is based upon the algorithm derived in `part_2.pdf`

```python
# Part 2.3:

# Import packages
import math
import numpy as np

# Our Binary tree approximation function:
def EDP(S0, T, r, sigma, M, f):
  """European Derivative Price binomial tree approximation of the provided problem
  Arguments:
    S0: initial stock price
    T: time horizon
    r: risk-free rate
    sigma: volatility
    M: number of nodes
    f: payoff function
  Returns:
    FP: Initial option value approximation
  """
  dt = T / M
  discount = math.exp(-r * dt)
  u = math.exp(r * dt) * (1 + math.sqrt(math.exp((sigma**2) * dt) - 1))
  d = math.exp(r * dt) * (1 - math.sqrt(math.exp((sigma**2) * dt) - 1))
  i = np.arange(0, M + 1, 1)
  TS = S0 * (u**i) * (d**(M - i))
  V = f(TS)
  for j in range(M):
    V = 0.5 * discount * (V[1:] + V[:-1])
  return V[0]
```

Use your code to price a derivative with payoff $$f(S)=\begin{cases}S-L & \text{if }S>K,\\0&\text{otherwise}\end{cases}$$ and the data $S_0 = £100$, $T = 1.5$, $r = 0.05$, $\sigma = 0.2$, $M = 500$, $K = S_0$ and $L = £90$.

```python
# Part 2.4

# References for Numpy methods used in my code:
# np.where(): element-wise conditional function
# eg: np.where(element-wise boolean condition, if true func, else func)
# https://numpy.org/doc/stable/reference/generated/numpy.where.html

# Define parameters:
S0 = 100
T = 1.5
r = 0.05
sigma = 0.2
M = 500

K = S0
L = 90

# Define our Payoff Function:
def payoff(S):
  return np.where(S > K, S - L, 0)

print("Inputs:\nS0 =", S0, ", T =", T, ", r =", r, ", sigma =", sigma, ",\nM =", M, ", K =", K, ", L =", L, "\n")
print(f"Initial price: {EDP(S0, T, r, sigma, M, payoff):.4f}")
```
